```python
DATASET_DIR = '../Dataset'
```

```python
COLUMNS = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
           'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
           '>50K']

NUMERICAL_COLS = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
CATEGORICAL_COLS = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 
                    'sex', 'native-country']

TARGET_COLUMN = '>50K'
```

```python
train_df = pd.read_csv(os.path.join(DATASET_DIR, 'adult.data'), names=COLUMNS, sep=',\s', na_values=['?'], engine='python')
train_df.head()
```

```python
test_df = pd.read_csv(os.path.join(DATASET_DIR, 'adult.test'), names=COLUMNS, sep=',\s', na_values=['?'], engine='python')
test_df.head()
```

```python
df = pd.concat((train_df, test_df))
df = df.dropna()
```

```python
df[TARGET_COLUMN] = df[TARGET_COLUMN].replace({'<=50K': 0, '>50K': 1})
```

## Split train / validation

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=5000, random_state=SEED)
```

```python
for train_index, test_index in sss.split(df.drop(TARGET_COLUMN, axis=1), df[TARGET_COLUMN]):
    train_df = df.iloc[train_index]
    dev_df = df.iloc[test_index]
```

## Save data

```python
ds_info = {
    'columns': COLUMNS,
    'categorical_columns': CATEGORICAL_COLS,
    'numerical_columns': NUMERICAL_COLS,
    'target_column': TARGET_COLUMN
}
```

```python
_ = joblib.dump(ds_info, os.path.join(DATASET_DIR, 'ds_info.jl'))
```

```python
train_df.to_csv(os.path.join(DATASET_DIR, 'n_train.csv'))
dev_df.to_csv(os.path.join(DATASET_DIR, 'n_dev.csv'))
```
